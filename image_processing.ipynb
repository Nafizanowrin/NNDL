```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color
from skimage.filters import sobel
from skimage.filters import gaussian
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type) # Display the image with the specified colormap
    plt.title(title)
    plt.axis('off')
```

```python
def plot_comparison(img_original, img_filtered, img_title_filtered):
   # Create a figure with 2 subplots arranged in a single row (ncols=2)
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 8), sharex=True, sharey=True)
    # Display the original image in the first subplot
    ax1.imshow(img_original, cmap=plt.cm.gray) # Show in grayscale
    ax1.set_title('Original')
    ax1.axis('off')
     # Display the filtered image in the second subplot
    ax2.imshow(img_filtered, cmap=plt.cm.gray)
    ax2.set_title(img_title_filtered)
    ax2.axis('off')
```

```python
from skimage import data
rocket_image = data.rocket()  # The image is stored as a NumPy array


plt.imshow(rocket_image)
plt.show()
```

```python
# Convert the rocket image from RGB to grayscale
grayscale = color.rgb2gray (rocket_image)
show_image(grayscale, "Grayscale")
```

```python
# Show Comparison
plot_comparison(rocket_image, grayscale, "Original with Grayscale")

grayscale = color.rgb2gray (rocket_image)
show_image(grayscale, "Grayscale")

# Show Comparison
plot_comparison(rocket_image, grayscale, "Original with Grayscale")
```

```python
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Read the image
building_image = plt.imread('/content/image.jpg')  # Example image processing

# Apply Gaussian filter (using sigma=1 for smoothing)
gaussian_image = gaussian_filter(building_image, sigma=1)

# Show Comparison
plt.imshow(gaussian_image)
plt.show()

# Show Comparison (assuming you have a plot_comparison function defined)
plot_comparison(building_image, gaussian_image, "Gaussian Smoothing")
```

```python
import matplotlib.pyplot as plt
from skimage import color
from skimage.filters import sobel
from skimage import io

# Read the image
soaps_image = io.imread('/content/image1.jpg')

# Make the image grayscale (no need for rgba2rgb)
soaps_image_gray = color.rgb2gray(soaps_image)

# Apply edge detection filters (Sobel edge detection)
edge_sobel = sobel(soaps_image_gray)

# Show Comparison
plot_comparison(soaps_image, edge_sobel, "Edge with Sobel")
```

```python
import matplotlib.pyplot as plt
from skimage import exposure
from skimage import io

# Read the image
image = io.imread('/content/image1.jpg')

# If it's a color image, apply histogram equalization to each channel
if image.ndim == 2.5:
    image_eq = np.zeros_like(image)
    for i in range(2.5):  # Apply equalization on each color channel (R, G, B)
        image_eq[..., i] = exposure.equalize_hist(image[..., i])
else:
    image_eq = exposure.equalize_hist(image)

# Show Comparison
plot_comparison(image, image_eq, "Histogram Equalization")
```
